# Ciclos en python

### 1) Ciclo While

El ciclo while repite una serie de instrucciones mientra la condición a evaluar sea verdadera

Sintaxis ciclo while

```python
# Sintaxis ciclo while
while condicion:
    # Bloque de código a ejecutar
```

se ejecuta hasta que la condicion sea falsa

Ejemplo:

````python
#imprimir los valores del 1 al 5

contador = 1

while contador <= 5:
    print(contador, end=' ')
    contador += 1 #contador = contador 

````

## 2) Ciclo for

El ciclo for itera o recorre una secuencia de valores, por ejemplo los caracteres de una cadena, una lista, etc, y ejecuta un bloque de código por cada elemento de la lista.

```python

#Sintaxis ciclo for
for variable in secuencia:
    #bloque de código a ejecutar


Ejemplo:

````python
#Ejemplo

cadena = 'Hola mundo'

for letra in cadena:
    print(letra, end=' ')
````

### 2.1) Cajero Automatico

Crear un cajero con las opciones:

1. Consultar saldo
2. Retirar
3. Depositar 
4. Salir



````python
#Primero se crea una variable constante 

SALDO_INICIAL = 1000

salir = False
print(f'*** Aplicación de Cajero Automático')
while not salir:
    print(f'''
    Operaciones que puedes realizar:
          1. Consultar Saldo
          2. Retirar
          3. Depositar
          4. Salir
    ''')
    opcion = int(input('Escoje una opción: '))

    if opcion == 1:
        print(f'Tu saldo actual es: ${SALDO_INICIAL:.2f}  \n')

    elif opcion == 2:
        retirar = float(input('Ingresa la cantidad que quiere retirar: '))

        if 0<=retirar <= SALDO_INICIAL:
            SALDO_INICIAL -= retirar
            print(f'Retira ${retirar:.2f}')
            print(f'Nuevo saldo actual: ${SALDO_INICIAL:.2f} \n')
        else:
            print('No cuenta con suficente saldo o ingreso una cantidad erronea \n')

    elif opcion == 3:
        deposito = float(input('Ingrese la cantidad a depositar: '))
        print(f'La cantidad a depositar es de ${deposito: .2f}\n')

        SALDO_INICIAL += deposito

        print(f'Su nuevo saldo es de ${SALDO_INICIAL: .2f}\n')

    elif opcion == 4:
        print('Cerrando sesión. Vuelva pronto.')
        salir = True

    
````

### 2.3) Juego de Adivinanzas

Crear un juego donde el jugador debe adivinar un número secreto.
Usando la funcion ranindt para generar un valor aleatorio entre 1 y 50.
Indicando si el numero fue mayor o menor al número secreto


````python
from random import randint

print('** Juego de Adivinanzas **')

numero_secreto = randint(1,50)

intentos = 0
INTENTOS_MAX = 5
adivinanza = None

while adivinanza != numero_secreto and intentos< INTENTOS_MAX:
    adivinanza = int(input('Adivina un número entre 1 y 50: '))

    # agregamos una ayuda
    if adivinanza > numero_secreto:
        print('El número ingresado es mayor al número secreto')

    else:
        print('El número ingresado es menor al número secreto')

    intentos += 1

#Coclusion dle juego modificada

if adivinanza == numero_secreto:
    print(f'Felicidades, adivinaste el número secreto en {intentos} intentos')

else:
    print(f'Lo sentimos agotaste el maximo de intentos: {INTENTOS_MAX}')
    print(f'El número secreto era: {numero_secreto}')

````

## 3) Función range en Python

Funcion range genera una secuencia de números

```python
# inicio - valor inicial
# fin - valor final, sin incluirlo
# incremento - diferencia entre cada número 
range(inicio, fin, incremento)

```

````python
# Ejemplo

for i in range(5):
    print(i, end=' ')

print('')
# imprimir del 0 al 9 con incrementos de 2

for i in range(0,10,2):
    print(i, end=' ')
````

### 3.1) Imprimir mensaje

````python
print('*** Repetición de un mensaje ***')

mensaje = input('Proporciona un mensaje a repetir: ')

numero_de_repeticiones = int(input('Proporciona el número de repeticiones: '))

#Ahora se itera sobre el rango de repeticiones

for i in range(numero_de_repeticiones):

    print(f'{i+1} - {mensaje}')
````

### 3.2) Triangulo simetrico}

````python
print('*** Triangulo simetrico ***')

numero_filas = int(input('Ingresa el número de filas: '))

for fila in range(1,numero_filas+1):
    espacios_blanco = ' ' * (numero_filas - fila)
    asteriscos = '*' * (2*fila - 1)

    print(f'{espacios_blanco}{asteriscos}')


````

## 4) Break y continue

````python
print('Break y continue')

#ejemplo con break

for numero in range(1,10):
    if numero % 2 == 0:
        print(numero)
        break ##Solo se muestra el primer numero par
````

````python
#Ejemplo con continue

for numero in range(1,10):
    if numero % 2 == 1:
        continue # en ves de terminar el ciclo lo que hace es realizar o revisar la siguiente iteración
    
    print(numero)
````
