# 1) Colecciones en python

Una colección es un conjunto de datos, con el objetivo de almacenar, organizar y manipular múltiples conjuntos de datos, conocidos como **Estructuras de datos**:

* Listas
* Tuplas
* Set (conjunto)
* Diccionarios


## 1) Listas

Las listas son ordenadas, mutables y son dinamicas (cambiar de tañano, añadir, modificar o eliminar)

```python 
#Sintaxis listas

mi_lista = [elemento1, elemento2, elemento3]
```

````python
# Ejemplo de listas

numeros = [1,2,3,4,5]
frutas = ['manzana', 'banana', 'cereza']
mixta = [1,'dos', 3.0, [4,5]]
````

````python
mi_lista = [1,2,3,4,5]

print(f'{mi_lista} - Lista original')


#Largo de la lista

print(f'Largo de la lista: {len(mi_lista)}')
````

````python
#Acceder a los elementos de la lista por indices

print(f'Accedemos al valor del indice 4: {mi_lista[4]}')


print(f'Accedemos al último indice de la lista: {mi_lista[-1]}')
````

````python
# Modificar los elementos de una lista 

mi_lista[1] = 10

print(f'Lista modificada : {mi_lista}')
````

````python
# Agregamos un nuevo elemento al final de la lista

mi_lista.append(6)

print(f'{mi_lista} - Se agregó el elemnto 6')
````

````python
# Anadir un nuevo elemento en un indice especifico
#lista.insert(indice, elemento)
mi_lista.insert(2,13)

print(f'{mi_lista} -- Se añadió el valor de 13 en el índice 2')
````

````python
#Eliminar elementos de una lista usando el método remove

mi_lista.remove(13)
print(f'{mi_lista} -- Se removio el elemento 13')
````

````python
#Removemos por indice con el metodo pop

mi_lista.pop(0)

print(f'{mi_lista} - se quito el elemento del indice 0')
````

````python
# Eliminar usando la palabra del

del mi_lista[2]

print(f'{mi_lista}--Se eliminó el índice 2')
````

````python
# Obtener sublistas

sublista = mi_lista[0:2] # genera una sublista del indice 0 al 1, el 2 no se incluye

print(f'{sublista}-- Se generó una sublista del indice 0 al 1 de la lista: {mi_lista}')
````

````python
#Ahora iteramos sobre una lista

print(f'** Iterar listas **')

nombres = ['Hugo', 'Vanessa', 'Fernanda']

for nombre in nombres:
    print(nombre, end=' ')
````

````python
lista_heterogenea = [100,True, 'Hugo']

for elemento in lista_heterogenea:
    print(elemento)
````

## 1.1) Playlist de Canciones

````python
for cancion in lista_reproduccion:
    print(f'- {cancion}')

````

````python
#Ahora hacemos que se agregen las canciones

lista_reproduccion=[]

num_canciones = int(input('Cuantas canciones deseas agregar? '))

#Ahora iteramos cada elemento de la lsita agregando un nuevo elemento

for indice in range(num_canciones):
    cancion = input(f'Proporciona la canción {indice+1}: ')

    lista_reproduccion.append(cancion)

lista_reproduccion.sort()

# mostrar la lista de canciones

print(f'\n Lista de Reproducción en orden alfabético: ')


for cancion in lista_reproduccion:
    print(f'- {cancion}')

    
````

### 1.2) Sistema de promedio de calificaciones

````python
print('*** Promedio de Calificaciones ***')

#Variables

lista_calificaciones = []

numero_calificaciones = int(input('Proporciona el no. de calificaciones a evaluar: '))

for indice in range(numero_calificaciones):
    calificacion = float(input(f'Calificacion [{indice}] = '))
    lista_calificaciones.append(calificacion)



#Calculamos el promedio de las calificaciones
#con una sum(Iterable)

suma_calificaciones = sum(lista_calificaciones)

promedio = suma_calificaciones/numero_calificaciones

print(f'Las calificaciones proporcionadas son: {lista_calificaciones}')

print(f'El promedio de calificaciones es {promedio: .2f}')
````

## 2) Tuplas

Similar a una lista, osea una colección de datos ordenados pero son inmutables, osea no es posible modificar su tamaño, ni modificar, ni eliminar elementos

```python 

#Sintaxis 
mi_tupla = (elemto1, elemento2, elemento3)

mi_tupla_sin_parentesis = elemento1, elemento2, elemento3

````python
# Ejemplo de tuplas

tupla1 = 'Juan', 'Karla'
tupla2 = 10, #Tupla de un elemento, pero con coma para una
````

````python
# Se accede de la misma forma

coordenada = (3,5)

print(f'\n Coordenada en el eje x: {coordenada[0]}')
print(f'Coordenada en el eje y: {coordenada[1]}')
````

Desempaquetado de Tuplas

````python
producto = ('P001', 'Camisa', 20.00)

#Desempaquetado

id, descripción, precio = producto

#Imprimir los valores

print(f'Tupla completa: {producto}')

#valores independienetes ya desempaquetados

print(f'Producto: id = {id}, descripción = {descripción}, precio = {precio}')
````

````python
# Combinar tuplas con listas

#definir una lisata que almacena tuplas de productos

productos = [
    ('P001', 'Camiseta', 20.00),
    ('P002', 'Jeans', 30.00 ),
    ('P003', 'Sudadera', 40.00)
]

#Imprimir la información de cada producto 
# y además calculamos el precio total

precio_total = 0

print('Información de los productos: ')

for producto in productos:
    id, descripción, precio = producto
    print(f'Producto: id = {id}, descripción = {descripción}, precio = ${precio}')

    precio_total += precio # O accediendo al indice 2

print(f'Precio total de los productos: ${precio_total}')
````

## 3) Conjuntos (Sets) en python

Los conjuntos colección de datos no ordenados de elementos únicos.

```python
#Sintaxis de un set

mi_set = {elemento1, elemento2, elemento3}


#Ejemplo

set_a = {1,2,3,4,5}

set_b = {3,'Juan', True, 6.5}
```

````python
#Agregar elementos

mi_set = {1,2,3,4,5,4}

print(f'Mi set: {mi_set}')

mi_set.add(6)
mi_set.add(7)


print(f'Mi set nuevo: {mi_set}')

````

````python
#eliminar un elemento del conjunto

mi_set.remove(4)

print(f'Mi set modificado: {mi_set}')


for elemento in mi_set:
    print(elemento, end=' ')
````

````python
#Comprobar si hay algun elemento

print(f'\n Existe el valorde 1 en el set? {1 in mi_set}')

#Obtener la longitud del set

print(f'El largo o logitud de mi conjunto es {len(mi_set)}')


````

### Operaciones entre conjuntos

````python
print('** Operaciones con el Set ***')

a = {1,2,3,4}
b = {3,4,5,6}

union = a | b

print(f'Union a | b : {union}')


interseccion = a & b

print(f'Intersección de a & b: {interseccion}\n')


diferencia = a - b

print(f'Diferencia de a - b: {diferencia}')
````

````python
print('*** Lista de Suscritores **')

suscriptores = {'luisa@mail.com', 'marcos@mail.com', 'elena@mail.com'}

print(f'Lista de suscriptores inicial: {suscriptores}')

#Verficiamos si un nuevo suscriptor ya está en la lista

nuevo_suscriptor = 'marcos@mail.com'

if nuevo_suscriptor in suscriptores:
    print(f'El nuevo suscriptor ya está en la lista {nuevo_suscriptor}')

else:
    suscriptores.add(nuevo_suscriptor)
    print(f'el nuevo suscriptor se ha agregado a la lista {nuevo_suscriptor}')
````
