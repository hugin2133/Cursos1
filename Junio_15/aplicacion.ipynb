### 1) Aplicación de salud y fitnes

Aplicación de salud y fitnes que solicite lo siguiente:

* Nombre del usuario
* Pasos caminados en el día

Se definen constantes:

* META_PASOS_DIARIOS = 10000
* CALORIAS_POR_PASOS = 0.04 #valor aproximado en kilocalorias

Las calorias quemadas según los pasos caminados;

<span style="color: blue;">calorias_quemadas = pasos_diarios * CALORIAS_POR_PASO</span>

Y se verifia si se cumplio la meta de pasos diarios.

<span style="color: blue;">meta_alcanzada = pasos_diarios > =  META_PASOS_DIARIO</span>

```python
print('*** Aplicación de salud y fitness')

# Constantes

META_PASOS_DIARIO = 10000
CALORIAS_POR_PASO = 0.04 # valor aproximado son kilocalorias

# Se piden los valores al usuario

nombre_usuario = input('¿Cuál es tu nombre?: ')
pasos_diarios = int(input('¿Cuántos pasos has caminado hoy?: '))

# Se verifica si el usuario alcanzó la meta de pasos diarios

meta_alcanzada = pasos_diarios >=  META_PASOS_DIARIO

meta_alcanzada_txt = 'Si' if meta_alcanzada else 'No'

#Se calcula las calorias quemadas

calorias_quemadas = pasos_diarios * CALORIAS_POR_PASO

#Se muestra la información

print(f'\nUsuario: {nombre_usuario}')

print(f'Pasos dados hoy: {pasos_diarios}')

print(f'Calorías quemadas: {calorias_quemadas} kcal')

print(f'Meta de pasos diarios alcanzada: {meta_alcanzada_txt}')

print(f'La meta de pasos diarios es de: {META_PASOS_DIARIO} pasos')
```

### 2) Sistema Reserva Hotel


Perdir la siguiente información:

* Nombre del cliente
* Días de estadia en el hotel
* Cuarto con vista al mar

Tarifas:

* Cuarto sin vista al mar $150.5 por dia
* Cuarto con vista al mar $190.5 por día



```python
#Definir constantes
COSTO_VISTA_MAR = 190.5 #por dia
COSTO_SIN_VISTA = 150.5 #por dia

nombre_cliente = input('Nombre del cliente: ')
dias_estadia = int(input('Días de estadia: '))
con_vista_mar = input('Con vita al mar (Si/No)?: ')

# vista_mar = con_vista_mar_txt.strip().lower() == 'si'



costo_estadia = dias_estadia*COSTO_VISTA_MAR if con_vista_mar.strip().lower() == 'si' else dias_estadia*COSTO_SIN_VISTA

print('\n ********** Detalles de la reservación**********')

print(f'Cliente: {nombre_cliente}')
print(f'Días de estadia: {dias_estadia}')
print(f'Costo total: ${costo_estadia}')
print(f'Habitación con vista al mar: {con_vista_mar.strip().lower()}')
```

### 3) El mayor de 2 Números

Indica cual es el mayor de dos numeros.

Pedir dos numero enteros

Se comparan y se imprime el numero mayor

```python
print('Ingrasa dos números enteros')
n_uno = int(input('Ingresa el primer número entero: '))
n_dos = int(input('Ingresa el segundo número entero: '))

mayor_primero = n_uno > n_dos

if mayor_primero:
    print(f'El primer número mayor es: {n_uno}')

else:
    print(f'El segundo número mayor es: {n_dos}')
```

### 4) Identifica la estación del Año

Se solicita el valor de un mes ( valor numérico entre 1 y 12) e indicar la estación del año según lo siguiente:

* Meses 1,2,12 Invierno
* Meses 3,4,5 Primavera
* Meses 6,7,8 Verano
* Meses 9,10,11 Otoño
* Cualquier otro valor Estacion desconocida

```python
print('Indentifica la estación del año')

n_mes = int(input('Ingresa el valor numerico del mes: '))
print('Ingresa el valor numerico del mes: ')

if n_mes in [1,2,12]:
    estacion = 'Invierto'

elif n_mes in [3,4,5]:
    estacion = 'Primavera'

elif n_mes in [6,7,8]:
    estacion = 'Verano'

elif n_mes in [9,10,11]:
    estacion = 'Otoño'

else:
    estacion = 'Estación desconocida'

print(f'La estación del año para el mes {n_mes} es {estacion}')
```

### 5) Sistema de Calificaciones

Calificación numérica (entre 0 y 10) a una letra (De la F a la A)

* Menor a 10 y mayor o igual a 9 es una A
* Menor a 9 y mayor o igual a 8 es una B
* Menor a 8 y mayor o igual a 7 es una C
* Menor a 7 y mayor o igual a 6 es una D
* Menor a 6 y mayor o igual a 0 es una F
* Otro caso Valor desconocido

```python
print('Sistema de calificaciones')

calificacion = float(input('Ingresa la calificacion del 0 al 10: '))

if 9<=calificacion<=10:
    letra = 'A'

elif 8<= calificacion <9:
    letra = 'B'

elif 7<= calificacion <8:
    letra='C'

elif 6<= calificacion <7:
    letra = 'D'

elif 0<= calificacion<6:
    letra = 'F'

else:
    letra = 'Desconocida'

print(f'Para la calificacion de  {calificacion} se le asigna la letra {letra}')
```
